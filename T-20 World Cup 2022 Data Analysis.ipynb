# Team Performance

```python
# Team Performance (assuming you have information about home_team and away_team)
team_runs = t20_data.groupby('home_team')['runs'].sum() + t20_data.groupby('away_team')['runs'].sum()
ax = team_runs.plot(kind='bar', title='Total Runs Scored by Teams')
plt.xlabel('Team')
plt.ylabel('Total Runs')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility

# Add labels on top of the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```

Srilanka Team Scored highest runs: 2218 and Scotland Team scored lowest runs:0

```python
team_wickets = t20_data.groupby('home_team')['wicket_id'].count() + t20_data.groupby('away_team')['wicket_id'].count()
ax = team_wickets.plot(kind='bar', title='Total Wickets Taken by Teams')
plt.xlabel('Team')
plt.ylabel('Total Wickets')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility

# Add labels on top of the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.tight_layout()  # Adjust layout for better spacing
plt.show()
plt.show()
```

Netherlands Team taken highest wickets: 111  and Scotland Team taken lowest wickets:0

# Over-by-Over Progress Analysis

```python
# Over-by-Over Progress: Runs
over_by_over_runs = t20_data.groupby(['over', 'ball'])[['runs']].sum().reset_index()
plt.figure(figsize=(16, 6))
sns.lineplot(x='over', y='runs', data=over_by_over_runs)
plt.title('Over-by-Over Progress: Runs')
plt.xlabel('Over')
plt.ylabel('Runs')
plt.show()
```

At 14th Over runs scored higher

```python
# Over-by-Over Progress: Wickets
over_by_over_wickets = t20_data.groupby(['over', 'ball'])[['bowler1_wkts', 'bowler2_wkts']].sum().reset_index()
plt.figure(figsize=(16, 6))
sns.lineplot(x='over', y='bowler1_wkts', data=over_by_over_wickets, label='Bowler 1')
sns.lineplot(x='over', y='bowler2_wkts', data=over_by_over_wickets, label='Bowler 2')
plt.title('Over-by-Over Progress: Wickets')
plt.xlabel('Over')
plt.ylabel('Wickets')
plt.legend()
plt.show()
```

# Critical Match Events Analysis

```python
# Extract relevant columns for analysis
critical_events = t20_data[['isBoundary', 'isNoball', 'isWide', 'wicket_id']]
```

```python
# Count occurrences of critical events
event_counts = critical_events.apply(lambda x: x.value_counts()).T.stack()
```

```python
# Get the top 10 critical events
top_10_events = event_counts.sort_values(ascending=False).head(10)
```

```python
# Plot Bar Chart for Top 10 Critical Events with Label Counts
plt.figure(figsize=(12, 6))
ax = top_10_events.plot(kind='bar')

# Add label counts on top of each bar
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.title('Top 10 Critical Match Events')
plt.xlabel('Event Type')
plt.ylabel('Count')
plt.show()
```

```python
# Analyze the impact of wickets
wickets_count = t20_data['wicket_id'].count()
```

```python
# Plot Pie Chart for Wickets Distribution
plt.figure(figsize=(8, 8))
plt.pie([wickets_count, len(t20_data) - wickets_count], labels=['Wickets', 'Non-Wickets'], autopct='%1.1f%%', startangle=90, colors=['red', 'green'])
plt.title('Distribution of Wickets')
plt.show()
```

# Consistent run-scorers

```python
# Evaluate standout players based on total runs
players_runs = t20_data.groupby(['batsman1_name', 'batsman2_name'])[['batsman1_runs', 'batsman2_runs']].sum().sum(axis=1)
top_run_scorers = players_runs.sort_values(ascending=False).head(10)
```

```python
# Evaluate consistent run-scorers
consistent_scorers = players_runs[players_runs >= players_runs.mean()]
```

```python
# Plot Pie Chart for Consistent Run Scorers
plt.figure(figsize=(8, 8))
plt.pie([len(consistent_scorers), len(players_runs) - len(consistent_scorers)], labels=['Consistent Scorers', 'Others'], autopct='%1.1f%%', startangle=90, colors=['green', 'gray'])
plt.title('Consistency of Run Scorers')
plt.show()
```
